```python
# %% [markdown]
# # Disease Prediction - Data Exploration
#
# This notebook explores the dataset to understand its structure, features, and target variable.

# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# %%
# Load the datasets
train_data = pd.read_csv('../data/training.csv')
test_data = pd.read_csv('../data/testing.csv')

```

```python
# %%
# Display basic information
print("Training Data Shape:", train_data.shape)
print("Testing Data Shape:", test_data.shape)

```

```python
# %%
# Check the first 5 rows of the training data
train_data.head()

```

```python
# %%
# Check data types and missing values
print("Training Data Info:")
print(train_data.info())
print("\nMissing Values in Training Data:")
print(train_data.isnull().sum())

```

```python
# %%
# Visualize disease distribution (using 'prognosis' instead of 'Disease')
plt.figure(figsize=(12, 6))
sns.countplot(y='prognosis', data=train_data, order=train_data['prognosis'].value_counts().index)
plt.title('Distribution of Diseases (Prognosis)')
plt.xlabel('Count')
plt.ylabel('Disease')
plt.tight_layout()
plt.savefig('../reports/visualizations/disease_distribution.png')
plt.show()
```

```python
# %%
# Check for class imbalance
disease_counts = train_data['prognosis'].value_counts()
print("\nDisease Counts:")
print(disease_counts)
```

```python
# %%
# Visualize symptom frequency (example: first 5 symptoms)
symptoms = train_data.columns[:-1]  # Exclude 'prognosis'
plt.figure(figsize=(12, 6))
for i, symptom in enumerate(symptoms[:5]):  # Plot first 5 symptoms
    plt.subplot(2, 3, i+1)
    sns.countplot(x=symptom, data=train_data)
    plt.title(symptom)
plt.tight_layout()
plt.show()
```
